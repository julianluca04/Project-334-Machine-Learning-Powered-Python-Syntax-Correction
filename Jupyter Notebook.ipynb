```python
def normalize_code(s):
    s = s.replace("\t", "    ")
    s = s.replace("\r\n", "\n").replace("\r", "\n")
    return s

df_1["buggy_chars"] = df_1["buggy_code"].astype(str).apply(normalize_code).apply(list)
df_1["fixed_chars"] = df_1["fixed_code"].astype(str).apply(normalize_code).apply(list)
```

```python
import re

def remove_noise(code):
    # remove Sample ID comments
    code = re.sub(r"#\s*Sample\s*ID\s*:\s*\d+", "", code)
    return code
```

```python
df_1["buggy_clean"] = (
    df_1["buggy_code"]
    .astype(str)
    .apply(remove_noise)
    .apply(normalize_code)
)

df_1["fixed_clean"] = (
    df_1["fixed_code"]
    .astype(str)
    .apply(remove_noise)
    .apply(normalize_code)
)

df_1["buggy_chars"] = df_1["buggy_clean"].apply(list)
df_1["fixed_chars"] = df_1["fixed_clean"].apply(list)
```
